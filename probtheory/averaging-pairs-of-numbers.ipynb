```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 100

x = np.arange(1, n+1)

means = np.zeros(n*n)

iteration = 0

for i in x:
    for j in x:
        means[iteration] = (i+j)/2
        iteration += 1

sample_size = 500
sample = np.random.choice(means, sample_size)
 
plt.hist(sample, 20)
plt.show()
```
