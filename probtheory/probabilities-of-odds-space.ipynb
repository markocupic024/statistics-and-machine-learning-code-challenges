```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
probs = np.zeros((11,10))

M = np.arange(1, 12)
N = np.arange(1, 11)

for i,m in enumerate(M):
    for j,n in enumerate(N):
        probs[i, j]=1/(1 + m/n)

plt.imshow(probs, cmap='hot', origin='lower')
plt.colorbar()
plt.show()
```
