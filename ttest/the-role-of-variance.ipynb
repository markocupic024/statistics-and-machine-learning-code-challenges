```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
n = 20
pop_mu = .5
iter = 100
sigmas = np.linspace(0.1,3,iter)

t_values = np.zeros(iter)
p_values = np.zeros(iter)

H0val = 0
df = n-1
```

```python
for idx, sigma in enumerate(sigmas):
    data = sigma * np.random.randn(n) + pop_mu
    t,p = stats.ttest_1samp(data,H0val)
    t_values[idx] = t
    p_values[idx] = p
```

```python
plt.plot(sigmas, t_values, 'or')
plt.title('t-values')
plt.show()
plt.plot(sigmas, p_values, 'o')
plt.title('p-values')
plt.show()
plt.plot(t_values, p_values, 'o')
plt.title('t-values by p-values')
plt.show()
```
