```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
n = 1000
thresh = 5

data_1 = np.exp(np.random.randn(n))
data_1[data_1>thresh] = thresh + np.random.randn(sum(data_1>thresh))*.1

data_2 = np.random.randn(n)

data = pd.DataFrame({'data_1':data_1, 'data_2':data_2})
data = data.melt()
data['origin'] = 0
sns.violinplot(data=data, y='value', split=True, hue='variable', x='origin')
```
