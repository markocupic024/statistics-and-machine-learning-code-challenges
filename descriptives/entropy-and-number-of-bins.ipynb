```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 1000

data = np.cumsum(np.sign(np.random.randn(N)))

bin_sizes = np.arange(4, 41)
entropies = np.zeros(len(bin_sizes))

for i, bin_size in enumerate(bin_sizes):
    values, _ = np.histogram(data, bin_size)
    probs = values/np.sum(values)

    entropy = -np.sum(probs*np.log2(probs + np.finfo(float).eps))
    entropies[i] = entropy
    
plt.plot(bin_sizes, entropies)
plt.xlabel("Number of bins")
plt.ylabel("Entropy (bits)")
plt.title("Using log2")
plt.show()

for i, bin_size in enumerate(bin_sizes):
    values, _ = np.histogram(data, bin_size)
    probs = values/np.sum(values)

    entropy = -np.sum(probs*np.log(probs + np.finfo(float).eps))
    entropies[i] = entropy

plt.plot(bin_sizes, entropies)
plt.xlabel("Number of bins")
plt.ylabel("Entropy (bits)")
plt.title("Using ln")
plt.show()
```
