```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 100

outlier = 20

bins = 20

data = np.append(np.random.randn(n), outlier)

y, x = np.histogram(data, bins)

x = (x[:-1] + x[1:]) / 2

mean = np.mean(data)

median = np.median(data)

print(mean)
print(median)

plt.plot(x, y)
plt.plot([mean,mean], [0,np.max(y)], label='Mean')
plt.plot([median,median], [0,np.max(y)], label='Median')
plt.legend()
plt.title('Small size, Small outlier')
plt.show()
```
