```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 1000

data = np.exp(np.random.randn(n)/2)

k = 40

fig, ax = plt.subplots(1, 2, figsize=(8,4))
ax[0].hist(data, bins=k)

y, x = np.histogram(data, bins=k)

y = 100*y/np.sum(y)

ax[1].hist(x[:-1], x, weights=y)
plt.show()
```
