```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0.001, 10, 100)

y1 = x

y2 = np.exp(x)

fig, ax = plt.subplots(1, 2, figsize=(8, 4))

ax[0].plot(x, y1)
ax[0].plot(x, y2)
ax[0].set_ylim([0,100])

ax[1].plot(x, y1)
ax[1].plot(x, y2)
ax[1].set_yscale("log")
```
